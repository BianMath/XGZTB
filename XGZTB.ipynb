```python
from docx import Document
import re
from IPython.display import clear_output
import tqdm
from time import sleep
```

```python
doc = Document("./《习近平新时代中国特色社会主义思想概论》习题集(20241031) 做题版.docx")
```

```python
tar_list = ['导    论',
 '第一章',
 '第二章',
 '第三章',
 '第四章',
 '第五章',
 '第六章',
 '第八章',
 '第九章',
 '第十章',
 '第十一章',
 '第十二章',
 '第十三章']
```

```python
res = []
temp = ""
index = 0
curr = tar_list[index]
for par in doc.paragraphs[21:]:
    if curr in par.text:
        index += 1
        try:
            curr = tar_list[index]
        except Exception as e:
            pass
        res.append(temp)
        temp = par.text
    else:
        temp += par.text + "\n"
res.append(temp)
res.pop(0)
res
```

```python
def judge(s):
    s = s[1:-1].strip()
    for item in s:
        if item not in "ABCDEF":
            return False
    return True
```

```python
title = "1.一个民族要走在时代前列就不能没有理论思维，一个国家要实现繁荣富强、人民幸福就不能没有科学理论指引。（ √ ）"
ans = re.findall("\（\s*√\s*\）|\（\s*×\s*\）", title)
print(ans)
```

```python
type_list = ["单项选择题", "多项选择题", "判断题", "简答题"]
# type_list = ["单选题", "多选题", "判断题", "简答题"]
text = res[11].split("\n")
index = 0
curr = type_list[index]
print(curr)
title = ""
res_dict = {}
for item in text:
    item = item.strip()
    if "本章参考答案" in item:
        break
    if curr in item:
        index += 1
        if index == 4:
            ans = re.findall("\（\s*√\s*\）|\（\s*×\s*\）", title)[0]
            title = re.sub(ans[1:-1].strip(), "", title)
            res_dict[title] = ans[1:-1].strip()
        title = ""
        try:
            curr = type_list[index]
            print(curr)
        except Exception as e:
            pass
        continue
    if index == 1 or index == 2:
        if item[0].isdecimal():
            title = ""
            title += item.strip()
        elif item[0] not in "ABCDEF":
            title += item.strip()
        elif item[0] == "A":
            ans = re.findall("\（\s*\w*\s*\）|\(\s*\w*\s*\)|\（\s*\w*\s*\)|\(\s*\w*\s*\）", title)
            for it in ans:
                if judge(it):
                    title = re.sub(it[1:-1].strip(), "", title)
                    correct = it[1:-1].strip()
                    break
            res_dict[title] = [correct, item.strip()]
        elif item[0] in "BCDEF":
            res_dict[title].append(item.strip())
    elif index == 3:
        if item[0].isdecimal():
            if title != "":
                ans = re.findall("\（\s*√\s*\）|\（\s*×\s*\）", title)[0]
                title = re.sub(ans[1:-1].strip(), "", title)
                res_dict[title] = ans[1:-1].strip()
            title = ""
            title += item.strip()
        else:
            title += item
    else:
        title += item      
res_dict["final"] = title
res_dict
```

```python
wrong_dic = {}
```

```python
%load_ext ipykernel
for key, values in tqdm.tqdm(res_dict.items()):
    if key == "final":
        print(values)
        con = input()
    elif type(values) == list:
        print("\n".join([key,*values[1:]]))
        sleep(0.5)
        ans = input("please enter:")
        print(f"right:{values[0]}")
        if ans.lower() != values[0].lower():
            wrong_dic[key] = values
        con = input()
        
    else:
        print(key)
        sleep(0.5)
        ans = input("please enter:")
        print(f"right:{values}")
        con = input()
    clear_output(wait=True)
```
